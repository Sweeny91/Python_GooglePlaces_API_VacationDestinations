```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
import json
import time

# Import API key
from api_keys import g_key
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
# Loading csv output from WeatherPy excercie
weatherdata_df = pd.read_csv("../output_data/cities.csv")
weatherdata_df.head(10)
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Configure gmaps
gmaps.configure(api_key=g_key)

# Store latitude and longitude lists in variable
locations = weatherdata_df[["Lat", "Lng"]]

# Store humidity list in variable
humidity = weatherdata_df["Humidity"]
intensity = weatherdata_df["Humidity"].max()
```

```python
# Plot Heatmap
fig_hm = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=float(intensity),
                                 point_radius=1)


# Add layer
fig_hm.add_layer(heat_layer)

# Display figure
fig_hm
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
# Ideal vacation weather conditions: '>80 degrees', 'wind speed 5 or <', 'humidity < 50', 'cloudiness < 30'
conditions = weatherdata_df[(weatherdata_df["Max Temp"] < 80) | (weatherdata_df["Wind Speed"] > 10) | 
                            (weatherdata_df["Humidity"] > 60) | (weatherdata_df["Cloudiness"] > 30)].index

weatherdata_df.drop(conditions, inplace = True)
weatherdata_df

# Save file for future reference
weatherdata_df.to_csv("../output_data/ideal_vac_locs.csv")
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
# Create list for hotel names to be appended to and specify radius
hotel_names = []
radius = 5000

# Loop through dataframe above to acquire locations
for row in range(len(weatherdata_df)):

    lats = weatherdata_df.iloc[row,1]
    lngs = weatherdata_df.iloc[row,2]
    coords = f"{lats},{lngs}"
    print(coords)

    params = {
        "location": coords,
        "types": "lodging",
        "radius": 5000,
        "key": g_key
    }

    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"   

    response = requests.get(base_url, params=params)
    vacspots_hotel = response.json()


    try:
        hotel_names.append(vacspots_hotel["results"][row]["name"])
    except:
        hotel_names.append("Nearest hotel not found")

print(hotel_names)
```

```python
# Create dataframe from these locations
weatherdata_df["Hotel Name"] = hotel_names
hotel_df = weatherdata_df
hotel_df["Date"] = pd.to_datetime(hotel_df["Date"])
hotel_df
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map
marker_locations = locations

# fig = gmaps.figure()
markers = gmaps.marker_layer(marker_locations,info_box_content = hotel_info)
fig_hm.add_layer(markers)

# Display figure
fig_hm
```
